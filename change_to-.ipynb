```python
def obtencion_de_nombre_png(path):
    files = []
    for i in os.listdir(path):
        if ".png" in i:
            files.append(i)
    return files
```

```python
#Obtenemos todos los archivos que hay en la carpeta del PATH
files = obtencion_de_nombre_png(path)
# files
```

```python
for file in files: 
    print(file)
    # new_new_file = Image.open(file)
    new_file = list(file)
    for i, v in enumerate(file):
        # print(i,v)
        if v  == "_":
            #Si alado del "_" hay un número, cambialo por "-".
            if new_file[i-1].isnumeric():
                new_file[i] = "-"
                new_string = "".join(new_file)
            else:
                new_string = file
    print(new_string)
    os.rename(os.path.join(path,file), os.path.join(path,new_string))
```
